```python
gplt.choropleth(
    indicadores,
    hue='estres_hidrico',
    cmap='coolwarm', linewidth=0.5,
    edgecolor='white',
    legend=True,
    legend_kwargs={'orientation': 'vertical'},
    figsize=(20,10),
    projection=gcrs.AlbersEqualArea()
)
plt.title("Estrs Hidrico (0-5)",fontsize=20,fontfamily='sans-serif')
plt.savefig("estres_hidrico.png", bbox_inches='tight', pad_inches=0.9)
```
