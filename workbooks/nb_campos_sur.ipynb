```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=df[df.capex_total <=200]
df1=df1.sort_values(by='tipo_formacion',ascending=False)

import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

plt.rcParams.update({'font.size': 40})
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Helvetica Neue']
plt.rcParams['font.style'] = 'normal'
plt.rcParams['font.stretch'] = 'normal'
plt.rcParams['font.weight'] = 'light'
plt.rcParams['axes.labelsize'] = 40
plt.rcParams['font.size'] = 40
plt.rcParams['legend.fontsize'] = 40
plt.rcParams['xtick.labelsize'] = 40
plt.rcParams['ytick.labelsize'] = 40
sns.set_palette('Set2')

fig, ax = plt.subplots(figsize=(95,45))
sns.scatterplot(x='capex_total', y='indice_vpn',
             hue='tipo_formacion',
             #size='ultimo_estado_reportado',
             #sizes=(1000,2000),
             hue_order=['Carbonatos','Areniscas'],
             alpha=0.6,
             legend='full',
             style="tipo_formacion",
             markers=True,
             data=df,s=4000,ax=ax)

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']-2, point['y']-1.5, str(point['val']),
                verticalalignment='top',horizontalalignment='left',
                ma='right',fontsize=40,linespacing=50,
               rotation_mode='anchor',rotation=0,wrap=True)

label_point(df1.capex_total, df1.indice_vpn, df1.campo, plt.gca()) 



plt.title(None)
# Set x-axis f
plt.xlabel(None)
# Set y-axis label
plt.ylabel(None)
plt.legend(loc='upper right',fontsize='large',markerscale=8., scatterpoints=1)

ax.set_xlim(0,200)

plt.savefig('scatter.png')
```
